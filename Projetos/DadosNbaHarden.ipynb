```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DataBase = pd.read_csv('Harden.csv')
```

```python
DataBase.head(
)
```

```python
DataBase.columns
```

```python
plt.figure(figsize=(16,8))
plt.title('Pontuação total James Harden por Temporada')
sns.barplot(data=DataBase,x='Season',y='PTS')
plt.xticks(rotation=45, ha='right');
```

```python
Data = DataBase.set_index('Season')

Data.head()
```

```python
rebotesTotais = DataBase['TRB'].sum()
```

```python
assistsTotais = DataBase['AST'].sum()
```

```python
#Mudar tabela para pts,rebotes e assists
plt.title('Evolução de Pontos ,Rebotes e Assistências durante a carreira de James Harden')
plt.plot(Data['EVOLUÇAO PTS POR ANO'],label='Pontuação Total')
plt.plot(assistsTotais,label='Assistencias Totais')
plt.plot(rebotesTotais,label='Rebotes Totais')
plt.legend()
plt.xticks(rotation=45, ha='right');
plt.show()
```
