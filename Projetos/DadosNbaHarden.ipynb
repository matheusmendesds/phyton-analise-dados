```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DataBase = pd.read_csv('Harden.csv')
```

```python
DataBase.head(
)
```

```python
DataBase.columns
```

```python
plt.figure(figsize=(16,8))
plt.title('Pontuação total James Harden por Temporada')
sns.barplot(data=DataBase,x='Season',y='PTS')
plt.xticks(rotation=45, ha='right');
```

```python
Data = DataBase.set_index('Season')

Data.head()
```

```python
# Totais
pontosTotais = DataBase['PTS'].sum()
assistsTotais = DataBase['AST'].sum()
rebotesTotais = DataBase['TRB'].sum()

# Dados para o gráfico
categorias = ['Pontos', 'Assistências', 'Rebotes']
totais = [pontosTotais, assistsTotais, rebotesTotais]

# Gráfico de barras
plt.title('Totais de Pontos, Rebotes e Assistências - Carreira James Harden')
plt.bar(categorias, totais, color=['blue', 'green', 'orange'])

# Mostrar valores em cima das barras
for i, v in enumerate(totais):
    plt.text(i, v + (v*0.01), str(v), ha='center', fontweight='bold')

plt.show()
```
